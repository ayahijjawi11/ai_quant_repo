```python
import pandas as pd
import numpy as np

np.random.seed(42)

regions = {
    "Gaza": ["Gaza City", "Khan Younis", "Rafah", "Deir al-Balah", "North Gaza"],
    "West Bank": ["Ramallah", "Nablus", "Hebron", "Jenin", "Tulkarm"]
}

facility_types = ["Hospital", "Emergency Center", "Water Pump", "Residential"]

data = []

for _ in range(250):
    region = np.random.choice(list(regions.keys()))
    zone = np.random.choice(regions[region])
    facility = np.random.choice(facility_types)
    hour = np.random.randint(0, 24)

    if facility == "Hospital":
        demand = np.random.uniform(100, 160)
        priority = 4
        pump_capacity = 0

    elif facility == "Emergency Center":
        demand = np.random.uniform(70, 120)
        priority = 3
        pump_capacity = 0

    elif facility == "Water Pump":
        demand = np.random.uniform(60, 100)
        priority = 3
        pump_capacity = np.random.uniform(300, 800)

    else:
        demand = np.random.uniform(30, 80)
        priority = 1
        pump_capacity = 0

    outage_risk = (
        np.random.uniform(0.6, 0.95)
        if region == "Gaza"
        else np.random.uniform(0.2, 0.6)
    )

    data.append({
        "region": region,
        "zone": zone,
        "facility_type": facility,
        "hour": hour,
        "demand_mw": round(demand, 2),
        "outage_risk": round(outage_risk, 2),
        "priority_level": priority,
        "pump_capacity_m3": round(pump_capacity, 1)
    })

df = pd.DataFrame(data)
df.to_csv("electricity_crisis_palestine_with_water.csv", index=False)

print("Dataset with water pumps saved successfully!")
import pandas as pd
import numpy as np

np.random.seed(42)

regions = {
    "Gaza": ["Gaza City", "Khan Younis", "Rafah", "Deir al-Balah", "North Gaza"],
    "West Bank": ["Ramallah", "Nablus", "Hebron", "Jenin", "Tulkarm"]
}

facility_types = ["Hospital", "Emergency Center", "Water Pump", "Residential"]

data = []

for _ in range(250):
    region = np.random.choice(list(regions.keys()))
    zone = np.random.choice(regions[region])
    facility = np.random.choice(facility_types)
    hour = np.random.randint(0, 24)

    if facility == "Hospital":
        demand = np.random.uniform(100, 160)
        priority = 4
        pump_capacity = 0

    elif facility == "Emergency Center":
        demand = np.random.uniform(70, 120)
        priority = 3
        pump_capacity = 0

    elif facility == "Water Pump":
        demand = np.random.uniform(60, 100)
        priority = 3
        pump_capacity = np.random.uniform(300, 800)

    else:
        demand = np.random.uniform(30, 80)
        priority = 1
        pump_capacity = 0

    outage_risk = (
        np.random.uniform(0.6, 0.95)
        if region == "Gaza"
        else np.random.uniform(0.2, 0.6)
    )

    data.append({
        "region": region,
        "zone": zone,
        "facility_type": facility,
        "hour": hour,
        "demand_mw": round(demand, 2),
        "outage_risk": round(outage_risk, 2),
        "priority_level": priority,
        "pump_capacity_m3": round(pump_capacity, 1)
    })

df = pd.DataFrame(data)
df.to_csv("electricity_crisis_palestine_with_water.csv", index=False)

print("Dataset with water pumps saved successfully!")
```
